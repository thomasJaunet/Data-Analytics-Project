```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = 'listings_borough.csv'
data_borough = pd.read_csv(filename)
del data_borough['Unnamed: 0']
data_borough.info()
```

```python
y = data_borough['price']
reviews_nb = data_borough['number_of_reviews']

plt.scatter(reviews_nb, y)
plt.show()
```

```python
sum(y>500)
```

```python
data_borough_500 = data_borough[data_borough['price'] < 500]
y_500 = data_borough_500['price']
reviews_nb_500 = data_borough_500['number_of_reviews']
```

```python
plt.scatter(reviews_nb_500, y_500)
plt.show()
```

After this analysis, we can see that the number of reviews doesn't explain the price. We can drop this feature, it is useless for our model

```python
avail = data_borough['availability_365']

plt.scatter(avail,y)
plt.show()
```

```python
avail_500 = data_borough_500['availability_365']

plt.scatter(avail_500,y_500)
plt.show()
```

We conclude that availability is not an explanatory feature for the price

```python
r_m = data_borough['reviews_per_month']

plt.scatter(r_m,y)
plt.show()
```

```python
r_m_500 = data_borough_500['reviews_per_month']

plt.scatter(r_m_500,y_500)
plt.show()
```

Not explanatory either

```python
r_m = data_borough['minimum_nights']

plt.scatter(r_m,y)
plt.show()
```

```python
r_m_500 = data_borough_500['minimum_nights']

plt.scatter(r_m_500,y_500)
plt.show()
```

Not very explanatory
